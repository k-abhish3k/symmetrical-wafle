*We can think of cost function as C(W,B,Sr,Er)*
- W -> Neural n/w Weights
- B -> Neural n/w biases

**BackPropagation**

- Fundamentally, we want to know how the cost function results changes with respect to the weights in the network, so we can update the weights to minimize the cost function.

- Consider a neural network with only one neuron in each layer - 
- **O--------O---------O-------O**
-w1+b1--w2+b2--w3+b3--w4+b4
-C(w1,b1,w2,b2,w3,b3,w4,b4)

let's say we have L layers, then our notation becomes:
- **O--------O---------O-------O**
-L-n--------L-2----------L-1---------L

Focusing on these last two layers, let's define **z=wx+b**
- The applying an activation function we'll state: **a=g(z)**

- As x is taken as the input to the first input layer, after every activation layer the input to the next layer changes to a ( g(z))

- Thus, we have:
 - **zL = (wL)(aL-1) + bL** 
 - where, aL-1 = output of (L-1)th layer
 - **aL = g(zL)**
 - **Cost function(C) = square(aL-y)**

Now we check how sensitive is the cost function towards a change in w: dC/dwL
- According to the chain rule **dC/dwL = dC/daL . daL/dzL . dzL/dwL**

We can calculate the same for the bias terms: 
- **dC/dbL = dC/daL . daL/dzL . dzL/dbL**

Idea : We can use the gradient to go back through the network and adjust our weights & biases to minimize the o/p of the error vector on the last o/p layer.

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1','feature2']].values  # Note : capital x cuz multi-dim dataframe
```

```python
y = df['price'].values # Note : small y cuz 1-dim dataframe
```

```python
# X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,random_state = 1)
```

```python
X_train.shape
```

```python
X_test.shape
```

**Normalizing & Scaling the data**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# help(MinMaxScaler)
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
 X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

**Creating & Training the Model**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
#help(Sequential)
```

```python
model = Sequential()

model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))

#Final output node for prediction
model.add(Dense(1))

model.compile(optimizer = 'rmsprop', loss='mse')
```

***Note: Choosing an optimizer and loss*** - 
- Keep in mind the kind of problem you are trying to solve:
 - For a multi-class classification problem:
   - model.compile(optimizer='rmsprop...etc', loss='categorical_crossentropy', metrics=['accuracy'])
 - For a Binary classification:
   - model.compile(optimizer='rmsprop....etc', loss='binary_crossentropy', metrics=['accuracy'])
 - For a mean squared error regression problem:
   - model.compile(optimizer='rmsprop...etc', loss='mse')  

```python
model.fit(x=X_train, y=y_train, epochs=250)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

**Model Evaluation**

```python
model.evaluate(X_test,y_test, verbose=1)
```

```python
model.evaluate(X_train, y_train, verbose=1)
```

```python
test_predictions = model.predict(X_test)
```
